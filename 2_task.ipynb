## Описание признаков датасета

- **Name** — имя респондента
- **Gender** — пол (Male/Female)
- **Age** — возраст
- **City** — город проживания
- **Working Professional or Student** — является ли человек работающим специалистом или студентом
- **Profession** — конкретная профессия или специальность
- **Academic Pressure** — уровень академического давления (шкала 1–5)
- **Work Pressure** — уровень рабочего давления (шкала 1–5)
- **Work/Study Hours** — количество часов работы или учёбы в день
- **CGPA** — средний балл успеваемости (Cumulative Grade Point Average)
- **Study Satisfaction** — удовлетворённость учёбой (шкала 1–5)
- **Job Satisfaction** — удовлетворённость работой (шкала 1–5)
- **Sleep Duration** — продолжительность сна (категории: "Less than 5 hours" и т.д.)
- **Dietary Habits** — качество питания (Healthy / Moderate / Unhealthy)
- **Degree** — уровень образования
- **Have you ever had suicidal thoughts?** — наличие суицидальных мыслей (Yes/No)
- **Financial Stress** — уровень финансового стресса (шкала 1–5)
- **Family History of Mental Illness** — наличие психических заболеваний в семейном анамнезе (Yes/No)
- **Depression** — наличие депрессии (Yes/No) — **целевая переменная**
