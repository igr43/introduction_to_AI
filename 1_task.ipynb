## Задача 1. Детектор хайпа (Pandas)

```python
import pandas as pd
import numpy as np

# Генерация данных (как в условии)
np.random.seed(42)
data = {
    'video_id': [f'V{i:03d}' for i in range(1, 21)],
    'views_start': np.random.randint(1000, 10000, 20),
    'views_end': np.random.randint(3000, 50000, 20),
    'category': np.random.choice(['Юмор', 'IT', 'Еда'], 20)
}
df = pd.DataFrame(data)

#1. Рассчитать колонку growth_rate — процент роста просмотров за день
df['growth_rate'] = ((df['views_end'] - df['views_start']) / df['views_start']) * 100

#2. Отфильтровать видео с ростом более 200%
hype_videos = df[df['growth_rate'] > 200].sort_values(by='growth_rate', ascending=False)

#3. Подсчитать количество хайповых видео по категориям
category_hype_count = hype_videos['category'].value_counts().reset_index()
category_hype_count.columns = ['category', 'hype_count']

#4. Вывести ТОП-3 самых быстрорастущих видео
top_3 = df.nlargest(3, 'growth_rate')[['video_id', 'growth_rate']]


print("Рост просмотров")
print(df[['video_id', 'views_start', 'views_end', 'growth_rate']].round(1))

print("\nХайповые видео (>200%)")
print(hype_videos[['video_id', 'category', 'growth_rate']].round(1))

print("\nКоличество хайповых видео по категориям")
print(category_hype_count)

print("\nТОП-3 самых быстрорастущих видео")
print(top_3)
```

## Задача 2. Зрение беспилотника (Numpy)


```python
import numpy as np

# 1. Создать массив 20×20 из нулей (пустая дорога)
road = np.zeros((20, 20))

# 2. Заполнить средний столбец (индекс 10) единицами (разметка)
road[:, 10] = 1

# 3. Сдвинуть все единицы на 2 позиции влево (поворот)
road[:, 8] = road[:, 10]
road[:, 10] = 0

# 4. Делаем разметку прерывистой: обнуляем каждую вторую единицу
road[1::2, 8] = 0


print(road[:, 8])
```

## Задача 3. ИИ-фильтр плохих слов (Pandas + Текст)

```python
import pandas as pd


messages = [
"Привет всем участникам чата",
"Розыгрыш iPhone прямо сейчас!!!",
"Кто знает хороший курс по Python?",
"@admin помогите разобраться",
"КРИПТА только растет #bitcoin $$$",
"Отличная статья про машинное обучение",
"Розыгрыш* призов каждый день",
"Подскажите книгу для начинающих",
"Заработок на крипте легко",
"Спасибо за помощь",
"### Супер предложение ###",
"Изучаю pandas для анализа данных",
"Розыгрыш 1000$ для всех",
"Какой фреймворк выбрать для веба?",
"Крипта это будущее финансов",
]
df = pd.DataFrame({'message': messages})

# 1. Найти сообщения со словами "крипта" или "розыгрыш" (регистронезависимо), пометить как is_spam = True
df['is_spam'] = df['message'].str.contains('крипта|розыгрыш', case=False, regex=True)

# 2. Удалить спецсимволы @ , # , $ , * с помощью регулярных выражений
df['cleaned_message'] = df['message'].str.replace('[@#$*]', '', regex=True)

# 3. Создать колонку word_count с количеством слов в сообщении
df['word_count'] = df['cleaned_message'].str.split().str.len()

# 4. Оставить только сообщения с более чем 3 словами, которые не являются спамом
result = df[(df['is_spam'] == False) & (df['word_count'] > 3)]


print(result[['message', 'is_spam', 'word_count']].reset_index(drop=True))
```

## Задача 4. Генератор лиц (Numpy)

```python
import numpy as np

# 1. Создать тензор размером 10×64×64×3 (10 RGB изображений 64×64)
#2. Заполнить случайными числами из нормального распределения (μ=0, σ=1)
images = np.random.normal(loc=0.0, scale=1.0, size=(10, 64, 64, 3))


random_idx = 3
# 3. Выбрать одно случайное изображение и заполнить каждый 4-й ряд красным цветом [255, 0, 0]

images[random_idx, ::4, :, :] = [255, 0, 0]  # ::4 — каждая 4-я строка

# 4. Создать "среднее изображение" размером 64×64×3 как среднее арифметическое всех 10 изображений
mean_image = np.mean(images, axis=0)  


print("Размер тензора images:", images.shape)         
print("Размер mean_image:", mean_image.shape)         
print("Значение в строке 0, столбце 0, канал R (красный):", mean_image[0, 0, 0])
```

## Задача 5. Система рекомендаций (Pandas + Numpy)

```python
import pandas as pd
import numpy as np


np.random.seed(42)
users = [f'User_{i}' for i in range(1, 11)]
genres = ['Рок', 'Поп', 'Техно', 'Рэп']
data = np.random.randint(0, 100, size=(10, 4))
df = pd.DataFrame(data, columns=genres, index=users)

# 1. Преобразовать DataFrame в матрицу Numpy
matrix = df.to_numpy()

# 2. Нормировать данные: для каждого пользователя сумма прослушиваний по всем жанрам должна равняться 1
row_sums = matrix.sum(axis=1, keepdims=True)  # Суммы по строкам 
normalized_matrix = matrix / row_sums  

# 3. Вычислить расстояние между первым пользователем и всеми остальными (евклидово расстояние)
first_user = normalized_matrix[0]  
distances = np.sqrt(np.sum((normalized_matrix - first_user) ** 2, axis=1))  # Евклидово расстояние

# 4. Найти ID пользователя с максимально похожими предпочтениями (минимальное расстояние, но не сам пользователь)
distances[0] = np.inf  
most_similar_idx = np.argmin(distances)  
most_similar_user = users[most_similar_idx]  # ID пользователя
distance_to_most_similar_user = distances[most_similar_idx]

print(df)

print("\nНормированная матрица")
print(normalized_matrix.round(3))

print("\nРасстояния до первого пользователя")
for user, dist in zip(users, distances):
    print(f"{user}: {dist:.3f}")

print("\nПользователь с наиболее похожими предпочтениями и расстояние до него")
print(f"{most_similar_user} и расстояние до него = {distance_to_most_similar_user:.7f}")
```
